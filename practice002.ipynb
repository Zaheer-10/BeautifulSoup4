## Import Libraries

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = 'https://www.byalicelee.com/'

r = requests.get(url)

htmlContent = r.content

soup = BeautifulSoup(htmlContent , 'html.parser')

print(soup.prettify())
```

```python
title = soup.title
print(title.string)
```

```python
paras = soup.find_all('div')
print("pragraphs : \n" , paras)
```
